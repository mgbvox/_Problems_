```python
def divisors(n):
    result = []
    for i in range(1,n):
        if n%i==0:
            result.append(i)
    return result

def is_prime(n):
    if n<1:
        return False
    for i in range(2,n):
        if n%i==0:
            return False
    return True

def prime_factors(n):
    if is_prime(n):
        return [n]
    else:
        largest_pdiv = [n for n in divisors(n) if is_prime(n)].pop()
        new_n = int(n/largest_pdiv)
        return [largest_pdiv] + prime_factors(new_n)
```

```python
'''
Shows non-linearity of the prime_factors function:
'''

import time
from tqdm import tqdm
exec_times = []
for i in tqdm(range(1,10000)):
    start_time = time.time()
    prime_factors(i)
    end_time = time.time()
    exec_times.append(end_time-start_time)
    
plt.plot(exec_times)
```
